```python
import random
import pandas as pd
import csv


data = pd.read_csv('./study.csv')
data
# data.sort_values(by=['우선순위 점수'])
```

```python
category_map = {'전공': 1.5, '교양': 1.0, '일반': 0.5}
data['가중치'] = data['구분'].map(category_map)
T = data['남은 시간']
P = data['학습 진도율']
C = data['가중치']
data['우선순위 점수'] = C * (1 / (T + 0.01) + (0.1 * (100 - P)))
data['우선순위 점수'] = data['우선순위 점수'].round(4)
data.to_csv('final.csv')
```
